```python
import networkx as nx

def generate_random_graph(n_nodes, n_edges):
    G = nx.connected_watts_strogatz_graph(n = n_nodes, k = n_edges, p=0.5)
    return G
```

```python
G = generate_random_graph(15,3)
pos =nx.spring_layout(G)
nx.draw_networkx_labels(G, pos)
nx.draw(G, pos)
```
